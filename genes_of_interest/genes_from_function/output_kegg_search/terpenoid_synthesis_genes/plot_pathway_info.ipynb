```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("terpenoid_synthesis__genes_of_interest.tsv",  sep='\t', index_col=0)
```

```python
pathway_ids = list()
pathway_names = list()
for full in df["PATHWAY"].tolist():
    for val in full.split(";"):
        if val.startswith("han"):
            pathway_ids.append(val)
        else:
            pathway_names.append(val)
# print(pathway_ids)
# print(pathway_names)
```

```python
from collections import Counter
```

```python
pathways_ids_count = Counter(pathway_ids)
pathways_names_count = Counter(pathway_names)
```

```python
pathways_names_count
```

```python
# plt = plt.fig()
plt.barh(list(pathways_ids_count.keys()), list(pathways_ids_count.values()))
plt.xlabel("Path ID")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.title("Pathways associated with all 88 terpenoid backbone synthesis genes (by KEGG pathway ID)")
```

```python
# plt = plt.fig()
plt.barh(list(pathways_names_count.keys()), list(pathways_names_count.values()))
plt.xlabel("Path ID")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.title("Pathways associated with all 88 terpenoid backbone synthesis genes (by pathway name)")
```
