```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

## No terpenoid gene expression, 2 different climate files

```python
no_terp_output = sorted(glob.glob("/Users/jennastanislaw/Boulder/team_rotation/TeamRotationSpring2025/sunflo_recode/examples_for_presentation/*noterp.csv"))
all_no_terp_dfs = dict()
for f in no_terp_output:
    name = f.split("/")[-1].split(".")[0]
    df = pd.read_csv(f, index_col=0)
    all_no_terp_dfs[name] = df
```
