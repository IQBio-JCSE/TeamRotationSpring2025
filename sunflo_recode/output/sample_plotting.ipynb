```python
all_dfs = list()
for f in all_output:
    df = pd.read_csv(f, index_col=0)
    all_dfs.append(df)
```

```python
plt.figure()
for df in all_dfs:
    plt.plot(df["CropBiomass"], label=df["a_water_stress"].tolist()[0])
plt.legend()
plt.show()
plt.savefig("varied_water_stress_biomassVthermaltime.png", dpi=400,  bbox_inches='tight')
```

```python
for column in df.columns:
    plt.figure()
    df[column].plot(title=column)
    plt.xlabel("Day")
    plt.ylabel(column)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure()
plt.scatter(df["ThermalTime"],df["CropBiomass"])
plt.title("Accumulation of crop biomass (g/m^2) with temperature")
plt.xlabel("ThermalTime")
plt.ylabel("Crop Biomass (g/m^2) ")
plt.savefig("test_biomassVthermaltime.png", dpi=400,  bbox_inches='tight')
```
